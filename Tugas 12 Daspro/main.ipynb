```python
import aritmatika.aritmatika as ar
import konversi.konversi as kv
import ubahBil.ubahBil as ub

def menuUtama():
    print ("""
    1. Aritmatika
    2. Konversi
    3. Ubah Bilangan
    4. Keluar
    """)

def menuAritmatika():
    print("""
    1. Penjumlahan
    2. Perpangkatan
    3. Perkalian
    """)
    
def menuKonversi():
    print ("""
    1. cm to meter
    2. meter to cm
    """)

def menuUbahBil():
    print ("""
    1. Desimal to Biner
    2. Desimal to Oktal
    3. Desimal to Hexadesimal
    """)

while True:
    menuUtama()
    pilihMenu = int (input ("Isikan Pilihan Menu:"))
    
    if pilihMenu == 1:
        menuAritmatika()
        pilihAritmatika = int (input ("Isikan Pilihan Perhitungan:"))
        if pilihAritmatika ==1:
            bil1 = int(input("Isikan Bilangan 1:"))
            bil2 = int(input("Isikan Bilangan 2:"))
            resultT = ar.penjumlahan(bil1,bil2)
            print(f"Hasil penjumlahan: {bil1} + {bil2} = {resultT}")
        elif pilihAritmatika ==2:
            bil1 = int(input("Isikan Bilangan 1:"))
            bil2 = int(input("Isikan Bilangan 2:"))
            resultP = ar.perpangkatan(bil1,bil2)
            print(f"Hasil Perpangkatan: {bil1} ^ {bil2} = {resultP}")
        elif pilihAritmatika ==3:
            bil1 = int(input("Isikan Bilangan 1:"))
            bil2 = int(input("Isikan Bilangan 2:"))
            resultK = ar.perkalian(bil1,bil2)
            print(f"Hasil Perkalian: {bil1} * {bil2} = {resultK}")
        else:
            print("Pilihan tidak tersedia")
            print("Mohon pilih dengan benar!!")
            continue
            
    elif pilihMenu == 2:
        menuKonversi()
        pilihKonversi = int(input("Isikan Pilihan Konversi:")) 
        if pilihKonversi ==1:
            pjg = float (input("Isikan panjang benda dalam sentimeter (bisa desimal):").replace(",", "."))
            resultCT = kv.cmToMeter(pjg)
            print(f"Hasil Konversi {pjg} cm ke {resultCT} meter")
        elif pilihKonversi ==2:
            pjg = float (input("Isikan panjang benda dalam Meter (bisa desimal):").replace(",", "."))
            resultTC = kv.meterToCm(pjg)
            print(f"Hasil Konversi {pjg} meter ke {resultTC} cm")
        else:
            print("Pilihan tidak tersedia")
            print("Mohon pilih dengan benar!!")
            continue
            
    elif pilihMenu == 3:
        menuUbahBil()
        pilihUbahBil = int (input ("Isikan Pilihan Ubah Bilangan:"))

        if pilihUbahBil ==1:
            bil = int(input("Isikan bilangan desimal:"))
            resultDB = ub.desimalToBiner(bil)
            print(f"Hasil Konversi {bil} dalam Biner = {resultDB}")
        elif pilihUbahBil ==2:
            bil = int(input("Isikan bilangan desimal:"))
            resultDO = ub.desimalToOktal(bil)
            print(f"Hasil Konversi {bil} dalam Oktal = {resultDO}")
        elif pilihUbahBil ==3:
            bil = int(input("Isikan bilangan desimal:"))
            resultDH = ub.desimalToHexadesimal(bil)
            print(f"Hasil Konversi {bil} dalam Hexadesimal = {resultDH}")
        else:
            print("Pilihan tidak tersedia")
            print("Mohon pilih dengan benar!!")
            continue
            
    elif pilihMenu == 4:
        print("Thank you!")
        break
```
