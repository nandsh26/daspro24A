```python
dataMahasiswa = []

while True:
    print("\n=== MENU MAHASISWA ===")
    print("1. Tambah Data Mahasiswa")
    print("2. Tampilkan Semua Data Mahasiswa")
    print("3. Cari Data Mahasiswa (berdasarkan NIM)")
    print("4. Keluar")

    pilihan = input("Pilih menu (1/2/3/4): ")

    if pilihan == '1':
        try:
            nim = input("Masukkan NIM (11 digit): ")
            nama = input("Masukkan Nama (min 6 karakter): ")
            alamat = input("Masukkan Alamat (min 10 karakter): ")
            kelas = input("Masukkan Kelas (TI23/TI24): ")

            if len(nim) != 11 or not nim.isdigit():
                raise ValueError("NIM harus 11 digit angka.")

            if len(nama.strip()) < 6:
                raise ValueError("Nama minimal 6 karakter.")

            if len(alamat.strip()) < 10:
                raise ValueError("Alamat minimal 10 karakter.")

            if kelas.upper() not in ['TI23', 'TI24']:
                raise ValueError("Kelas harus TI23 atau TI24.")

            dataMahasiswa.append([nim, nama, alamat, kelas])

            dataMahasiswa.sort() 
            dataPertama = dataMahasiswa[0]
            dataKeenam = dataMahasiswa[5] 

        except IndexError:
            print("Salah mengakses Indeks (data kurang dari 6)")
        except ValueError as ve:
            print(f"Kesalahan input: {ve}")
        else:
            print("\nData berhasil ditambahkan!")
            print(f"Data Pertama (berdasarkan NIM): {dataPertama}")
            print(f"Data Keenam: {dataKeenam}")
        finally:
            print("Proses input selesai.\n")

    elif pilihan == '2':
        try:
            if len(dataMahasiswa) == 0:
                raise IndexError("Belum ada data yang dimasukkan.")
            
            print("\n=== DAFTAR MAHASISWA ===")
            for i, mhs in enumerate(dataMahasiswa, 1):
                print(f"{i}. NIM: {mhs[0]}, Nama: {mhs[1]}, Alamat: {mhs[2]}, Kelas: {mhs[3]}")
        except IndexError as e:
            print(f"Error: {e}")
        finally:
            print("Selesai menampilkan data.\n")

    elif pilihan == '3':
        try:
            if len(dataMahasiswa) == 0:
                raise IndexError("Tidak ada data untuk dicari.")
            
            cari_nim = input("Masukkan NIM yang ingin dicari: ")
            ditemukan = False

            for mhs in dataMahasiswa:
                if mhs[0] == cari_nim:
                    print(f"Data ditemukan: NIM: {mhs[0]}, Nama: {mhs[1]}, Alamat: {mhs[2]}, Kelas: {mhs[3]}")
                    ditemukan = True
                    break

            if not ditemukan:
                print("Data dengan NIM tersebut tidak ditemukan.")

        except IndexError as e:
            print(f"Error: {e}")
        finally:
            print("Proses pencarian selesai.\n")

    elif pilihan == '4':
        print("Terima kasih, program selesai.")
        break
    else:
        print("Pilihan tidak valid. Coba lagi.")
```
